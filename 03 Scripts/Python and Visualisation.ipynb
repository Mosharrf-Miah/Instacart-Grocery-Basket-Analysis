#### Creating charts - Bar chart

```python
ords_prods_merge['orders_day_of_the_week'].value_counts().plot.bar()
plt.title('Order Days Of The Week')
plt.ylabel('Number of Orders (Millions)')
plt.xlabel('Days Of The Week')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
```

```python
## changing the sort of the bar chart
```

```python
ords_prods_merge['orders_day_of_the_week'].value_counts().sort_index().plot.bar()
plt.title('Order Days Of The Week')
plt.ylabel('Number of Orders (Millions)')
plt.xlabel('Days Of The Week')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.savefig('bar_orders_dow', dpi=300, bbox_inches='tight')
```

## Merging both datasets

```python
## the product_id column is present in both datasets allowing us to merge both the datasets together. 
## an inner join will provide the best result in this scenario as it will provide the optimum outcome.
```

```python
## checking the combined dataset without creating the dataframe
```

```python
ords_prods_merge = df_orders_products_combined.merge(df_prepared_products, on = ['product_id'], indicator = 'Match')
```

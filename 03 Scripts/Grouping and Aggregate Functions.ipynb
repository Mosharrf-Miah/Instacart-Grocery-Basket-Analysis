## Aggregating data with Transform()

```python
## finding loyal customers using transform and max
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(ny.max)
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge['max_order'].value_counts()
```

#### Creating a new loyalty column with loc

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal_customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge['loyalty_flag']
```

#### Analysing Customer Spending Habits

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

#### Average price spend by each user

```python
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(ny.mean)
```

```python
ords_prods_merge[['user_id', 'avg_price']].head(100)
```

## Creating a spending flag to find high and low spenders

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] >= 10, 'spending flag'] = 'High spender'
```

```python
ords_prods_merge['spending flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head(50)
```

#### Creating Order Frequency Flag

```python
## creating a flag to determine user by frequent customer and non frequent customer
```

```python
ords_prods_merge['avg_days_between_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(ny.mean)
```

```python
ords_prods_merge['avg_days_between_order']
```

```python
ords_prods_merge['avg_days_between_order'].value_counts(dropna = False)
```

```python
## I have determined the average days between the last order for each customer, Now i will need to put this information into a flag
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_days_between_order'] > 20, 'order frequency flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['avg_days_between_order'] > 10) & (ords_prods_merge['avg_days_between_order'] <= 20), 'order frequency flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_days_between_order'] <= 10, 'order frequency flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order frequency flag'].value_counts(dropna = False)
```
